```python
import gymnasium as gym
import numpy as np
import random
```

### Gymnasium Environment Lifecycle

1. **Create** $\rightarrow$ `gym.make()`
2. **Reset** $\rightarrow$ `env.reset()`
3. **Action** $\rightarrow$ `env.step()`
4. **Show** $\rightarrow$ `env.render()`
5. **Close** $\rightarrow$ `env.close()`

```python
#sample environment
env = gym.make("CliffWalking-v1", render_mode="human")
```

```python
print(env.observation_space.n) # states
print(env.action_space.n) # actions
```

```python
gamma = 0.99
alpha = 0.5
espilon = 0.1
episodes = 500
```

```python
# Q-table => stores Q-values
Q=np.zeros((48,4))
```

```python
Q # actions can be performed for each state so 48x4
```

```python
# Policy -  epsilon greedy

def espilon_greedy(state):
    if random.random() < espilon:
        return env.action_space.sample() #random action (explore)
    else:
        return np.argmax(Q[state])       # exploit
```

```python
for episode in range(episodes):
    # render = (episode % 50 ==0 )
    # if render:
    #         env = gym.make("CliffWalking-v1",render_mode="human")
    # else:
    env = gym.make("CliffWalking-v1")
    done = False  # this tells when the episode has ended
    state,_ = env.reset()
    action = espilon_greedy(state)

    total_reward = 0
    episode_len = 0

    while not done:
        next_state, reward, terminated, truncated, _ = env.step(action)
        done = terminated or truncated
        next_action = espilon_greedy(next_state)

        #SARSA Update
        Q[state,action] += alpha*(reward+gamma*(Q[next_state,next_action]-Q[state,action]))
        state = next_state
        action = next_action

        total_reward +=reward
        episode_len+=1
    print(f"{episode+1}/500")
    print(f"total reward = {total_reward}")
    print(f"episode length = {episode_len}")
    env.close()
```

```python
env = gym.make("CliffWalking-v1" , render_mode="human")
state,_ = env.reset()
done = False
total_reward =0
episode_len=0

while not done:
    action = np.argmax(Q[state])
    next_state,reward,terminated,truncated,_ = env.step(action)
    done = terminated or truncated
    state = next_state
    total_reward +=reward
    episode_len += 1
print(f"{episode+1}/500")
print(f"total reward = {total_reward}")
print(f"episode length = {episode_len}")
env.close()
```
